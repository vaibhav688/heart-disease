```python
import pandas as pd
data=pd.read_csv('/kaggle/input/heart-disease-uci/heart.csv')
```

```python
inputs_col=list(data.columns)[0:-1]
targets_col='target'
```

```python
inputs=data[inputs_col]
target=data[targets_col]
```

```python
inputs
```

```python
import numpy as np
numeric_col=inputs.select_dtypes(include=np.number).columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean').fit(inputs)
```

```python
inputs[numeric_col]=imputer.transform(inputs[numeric_col])
```

```python
inputs[numeric_col]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler().fit(inputs[numeric_col])
```

```python
inputs[numeric_col]=scaler.transform(inputs[numeric_col])
```

```python
inputs.describe()
```

```python
from sklearn.model_selection import train_test_split
train_input,val_input,train_target,val_target=train_test_split(inputs,target,test_size=0.2,random_state=42)
```

```python
X_train=train_input.copy()
X_val=val_input.copy()
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression().fit(train_input,train_target)
```

```python
mod.score(X_train,train_target)
```

```python
mod.score(X_val,val_target)
```

```python
from sklearn.tree import DecisionTreeClassifier
tr=DecisionTreeClassifier(max_depth=3,random_state=42).fit(train_input,train_target)
```

```python
tr.score(X_train,train_target),tr.score(X_val,val_target)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_jobs=-1,max_depth=4,criterion='entropy',max_features=3,random_state=42).fit(train_input,train_target)
```
